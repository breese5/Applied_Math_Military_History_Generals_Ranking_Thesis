```python

%pip install Wikipedia-API

import wikipediaapi
import csv
import pandas as pd


wiki_wiki = wikipediaapi.Wikipedia('MilitaryGenerals (bryant.reese@yale.edu)','en')

def get_page_links(page):
    """ Extract links to other Wikipedia pages from a given page. """
    links = page.links
    return [link for link in links.values() if 'Battle of' in link.title]

def get_battle_description(page):
    """ Extract the first section (summary) of a battle page as its description. """
    return page.summary

def main():
    # Get the main list of battles page
    list_of_battles = wiki_wiki.page('List of battles (alphabetical)')
    
  
    battle_links = get_page_links(list_of_battles)

    battles_data = []
    
  
    for battle in battle_links:
        battle_page = wiki_wiki.page(battle.title)
        description = get_battle_description(battle_page)
        battles_data.append([battle.title, description])
        print(f'Processed: {battle.title}')  

    # Save data to CSV
    with open('battles_descriptions.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['Battle Name', 'Description'])
        writer.writerows(battles_data)
    

    df = pd.DataFrame(battles_data, columns=['Battle Name', 'Description'])
    return df


df_battles = main()
df_battles.head()
```
